```python
def clean_txt(text):
    search = ['أ','آ','إ','ة','ى']
    replace = ['ا','ا','ا','ه','ي']
    
    
    for i in range(len(search)):
        text = text.replace(search[i], replace[i])
    
    words = re.split("[^0-9a-zA-Zابتثجحخدزرزسشصضطظعغفقكلومنهةىؤئء,يأآئ]*", text)
    
    return " ".join(words).strip().lower()
```

```python
def filter_exp_years(text):
    #check if there are numbers
    text = text.replace("-", ' - ')
    text = text.replace("+", ' - ')
    text = text.replace("/", ' - ')
    text = text.replace("~", ' - ')
    text = word_tokenize(text)
    
    numbers = [int(s) for s in text if s.isdigit()]
    words = [s.lower().strip() for s in text if not s.isdigit()]
    
    per = 'year'
    if('month' in words or 'months' in words):
        per = 'month'
    if(len(numbers) == 1):
        if('less' in words or 'maximum' in words): 
            return (0,max(numbers), per)
        elif('more' in words or 'above' in words or 'minimum' in words):
            return (max(numbers),max(numbers), per)
        else:
            return (0,max(numbers), per)
    elif (len(numbers) > 1):
        return (min(numbers), max(numbers), per)
    else:
        return (0,0, per)
```

```python
err = 0
for i in range(df.shape[0]):
    # filter salary
    if(int(df.iloc[i]['salary_min']) <= 0 ):
        df.set_value(i, 'salary_min', 0)
    if(int(df.iloc[i]['salary_max']) <= 0 ):
        df.set_value(i, 'salary_max', 0)

    df.set_value(i, 'salary_avg', (df.iloc[i]['salary_max'] + df.iloc[i]['salary_min'])/2.0 )
    
    # filter num vacancies
    if(int(df.iloc[i]['num_vacancies']) <= 0 ):
        df.set_value(i, 'num_vacancies', 0)
    
    # filter exp years
    exp_yrs = filter_exp_years(str(df.iloc[i]['experience_years']))
    df.set_value(i, 'experience_years_min', exp_yrs[0])
    df.set_value(i, 'experience_years_max', exp_yrs[1])
    df.set_value(i, 'experience_years_type', exp_yrs[2])
    
    # parse date/time
    d = df.iloc[i]['post_date'].split('-')
    df.set_value(i, 'date_year', d[0])
    df.set_value(i, 'date_month', datetime.date(1900, int(d[1]), 1).strftime('%B'))
    
    # filter job_reqs
    df.set_value(i, 'clean_description', html2text.html2text(str(df.iloc[i]['description'])))
    df.set_value(i, 'clean_job_requirements', html2text.html2text(str(df.iloc[i]['job_requirements'])))
    
    # filter city_name
    df.set_value(i, 'city_name', clean_txt(df.iloc[i]['city_name']))
    
```

```python
# logistic regression to predict views
train_data = df_2[['experience_years_min', 'experience_years_max','date_month_codes','salary_avg',
                   'num_vacancies',
    'city_name_codes','job_industry_1_codes', 'job_category_1_codes', 'career_level_codes']]
train_cls = df_2['views']
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, train_cls, test_size=0.4, random_state=101)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_,train_data.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
sns.distplot((y_test-predictions),bins=50)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
